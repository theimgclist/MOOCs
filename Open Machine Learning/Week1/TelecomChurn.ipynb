```python
d = {'No' : False, 'Yes' : True}
```

```python
dataset['International plan'] = dataset['International plan'].map(d)
dataset.head()
```

```python
dataset = dataset.replace({'Voice mail plan' : d})
```

```python
dataset.head()
```

```python
columns_to_show = ['Total day minutes', 'Total eve minutes', 'Total night minutes']
dataset.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

```python
dataset.groupby(['Churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

```python
pd.crosstab(dataset['Churn'], dataset['International plan'])
```

```python
pd.crosstab(dataset['Churn'], dataset['Voice mail plan'], normalize=True)
```

```python
dataset.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc='mean')
```

```python
total_calls = dataset['Total day calls'] + dataset['Total eve calls'] +\
dataset['Total night calls'] + dataset['Total intl calls']
dataset.insert(loc = len(dataset.columns), column = 'Total calls',\
value=total_calls)
dataset.head()
```

```python
dataset['Total charge'] = dataset['Total day charge'] + dataset\
['Total eve charge'] + dataset['Total night charge'] + \
dataset['Total intl charge']
dataset.head()
```

```python
dataset.drop(['Total charge', 'Total calls'], axis = 1, inplace = True)
```

```python
dataset.shape
```

```python
pd.crosstab(dataset['Churn'],dataset['International plan'], margins = True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (8,6)
sns.countplot(x='International plan', hue='Churn', data=dataset);
```

```python
pd.crosstab(dataset['Churn'],dataset['Customer service calls'],margins=True)
```

```python
sns.countplot(x='Customer service calls', hue='Churn',data=dataset)
```

```python
dataset['Many service calls'] = (dataset['Customer service calls'] > 3).astype('int')
pd.crosstab(dataset['Many service calls'],dataset['Churn'], margins= True)
```

```python
sns.countplot(x='Many service calls', hue='Churn', data = dataset);
```

```python
# We can draw some helpful insights using the analysis done so far
# International plan and Customer Service Calls can help predicting the churn outcome
# International plan = True & Customer Service calls > 3 => Churn = 1, else Churn = 0
# The above simple condition can be a good starting point for the model or the baseline model
```
